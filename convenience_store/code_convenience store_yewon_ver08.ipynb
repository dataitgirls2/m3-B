지역명을 짧은 이름으로 매치시키기

```python
# 고성군은 강원도, 경상남도에 있으므로 ()안에 표시해주기
# 세종특별시는 세종으로 바꿔줌
# 인천옹진, 울산울주, 대구달성, 부산기장은 군이라서 따로 설정해줘야 함! 억지로 넣어버려 ㅇㅅㅇ!
def short_distr(name):
    wide, narrow = name.split()
    if narrow.endswith('구'):
        return wide[:2] + (narrow[:-1] if len(narrow) > 2 else narrow)
    elif narrow == '고성군':
        return '고성({})'.format({'강원도': '강원', '경상남도': '경남'}[wide])
    elif narrow == '세종특별자치시': 
        return '세종'
    elif narrow == '옹진군': 
        return '인천옹진'
    elif narrow == '울주군': 
        return '울산울주'
    elif narrow == '달성군': 
        return '대구달성'
    elif narrow == '기장군': 
        return '부산기장'
    else:
        return narrow[:-1]

convenience_df_sort['shortname'] = list(map(short_distr, convenience_df_sort.index))
convenience_df_sort['shortname'].unique()
```

```python
# 블록맵을 만들기 위해 draw_blockmap함수 생성
# draw_blockmap('column', df, 'd1', 'd2', 'color', np.arange(start, stop, step), 'title')
def draw_blockmap(targetData, blockedMap, d1, d2, cmapname, dataticks, title):
    gamma = 0.75

    whitelabelmin = (max(blockedMap[targetData]) - min(blockedMap[targetData])) * 0.5 + min(blockedMap[targetData])

    datalabel = targetData

    vmin = min(blockedMap[targetData])
    vmax = max(blockedMap[targetData])

    mapdata = blockedMap.pivot(index='y', columns='x', values=targetData)
    masked_mapdata = np.ma.masked_where(np.isnan(mapdata), mapdata)
    
    plt.figure(figsize=(9, 12))
    plt.pcolor(masked_mapdata, vmin=vmin, vmax=vmax, cmap=cmapname, edgecolor='#aaaaaa', linewidth=0.6)

    # 지역 이름 표시
    for idx, row in blockedMap.iterrows():
        annocolor = 'white' if row[targetData] > whitelabelmin else 'black'

        # 광역시는 구 이름이 겹치는 경우가 많아서 시단위 이름도 같이 표시한다. (중구, 서구)
        if row[d1].endswith('시') and not row[d1].startswith('세종'):
            dispname = '{}\n{}'.format(row[d1][:2], row[d2][:-1])
            if len(row[d2]) <= 2:
                dispname += row[d2][-1]
        elif row[d1].startswith('세종'):
            dispname = '세종'
        else:
            dispname = row[d2][:-1]

        # 서대문구, 서귀포시 같이 이름이 3자 이상인 경우에 작은 글자로 표시한다.
        if len(dispname.splitlines()[-1]) >= 3:
            fontsize, linespacing = 9.5, 1.2
        else:
            fontsize, linespacing = 12, 1.2

        plt.annotate(dispname, (row['x']+0.5, row['y']+0.5), weight='bold',
                     fontsize=12.6, ha='center', va='center', color=annocolor,
                     linespacing=linespacing)
        
    # 시도 경계 그린다.
    for path in BORDER_LINES:
        ys, xs = zip(*path)
        plt.plot(xs, ys, c='black', lw=1.2)

    plt.gca().invert_yaxis()
    #plt.gca().set_aspect(1)

    plt.axis('off')

    cb = plt.colorbar(shrink=.1, aspect=10)
    cb.set_label(datalabel)
    cb.set_ticks(dataticks)
    
    # title 생성
    plt.tight_layout()
    plt.title(title, y=1.02, loc='left', fontproperties='NanumBarunGothic', fontsize=18, fontweight='bold')

    plt.show()
```

```python
# 인구밀도와 편의점지수의 상관관계보기
validconv = df.dropna(subset=['편의점지수']).copy()
validconv['log편의점지수'] = np.log2(validconv['편의점지수'])
validconv = validconv.dropna(subset=['log편의점지수'])
validconv['log인구밀도'] = np.log10(validconv['인구밀도'])

fig = plt.figure(figsize=(4, 4))
ax = fig.add_subplot(1, 1, 1)
validconv.plot(kind='scatter', y='log편의점지수', x='log인구밀도', ax=ax,
             edgecolor='none', s=8, c='black')
plt.xlabel('인구밀도 (log$_{10}$ 인/km$^2$)')
plt.ylabel('편의점지수 (log$_2$)')

tau, taup = scipy.stats.kendalltau(validconv['log편의점지수'], validconv['log인구밀도'])
print("Kendall's tau: {} (p={})".format(tau, taup))
plt.annotate('$\\tau$ = {:.3f}'.format(tau), (2, 2), fontsize=14)
```

인구와 매장수

```python
GS_per_population = df['GS25'].sum() / df['인구수'].sum()
GS_to_random = df['GS25'] / (GS_per_population * df['인구수'])

CU_per_population = df['CU'].sum() / df['인구수'].sum()
CU_to_random = df['CU'] / (CU_per_population * df['인구수'])

EMART_per_population = df['EMART'].sum() / df['인구수'].sum()
EMART_to_random = df['EMART'] / (EMART_per_population * df['인구수'])

SEVEN_per_population = df['7ELEVEN'].sum() / df['인구수'].sum()
SEVEN_to_random = df['7ELEVEN'] / (SEVEN_per_population * df['인구수'])

MINI_per_population = df['MINISTOP'].sum() / df['인구수'].sum()
MINI_to_random = df['MINISTOP'] / (MINI_per_population * df['인구수'])
```

```python
import rpy2.robjects as ro

def loess_fit(x, y, px=None, model=None, alpha=0.5):
    if model is None:
        model = ro.r('y ~ x')

    if px is None:
        px = np.linspace(min(x), max(x), 22)[1:-1]

    fitframe = ro.DataFrame({'x': ro.FloatVector(x), 'y': ro.FloatVector(y)})
    loessmodel = ro.r.loess(model, fitframe, span=alpha)

    predframe = ro.DataFrame({'x': ro.FloatVector(px)})
    predy = ro.r.predict(loessmodel, predframe)
    preddata = [(x, predy[i]) for i, x in enumerate(px)]

    return np.array(preddata).transpose()

GS_trend = loess_fit(np.log10(df['인구수']), GS_to_random)
CU_trend = loess_fit(np.log10(df['인구수']), CU_to_random)
EMART_trend = loess_fit(np.log10(df['인구수']), EMART_to_random)
SEVEN_trend = loess_fit(np.log10(df['인구수']), SEVEN_to_random)
MINI_trend = loess_fit(np.log10(df['인구수']), MINI_to_random)
```
